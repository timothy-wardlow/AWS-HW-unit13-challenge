# Clustering Crypto

```python
# Initial imports
import requests
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import hvplot.pandas
from sklearn.preprocessing import StandardScaler, MinMaxScaler
from sklearn.decomposition import PCA
from sklearn.cluster import KMeans
```

### Fetching Cryptocurrency Data

```python
# Use the following endpoint to fetch json data
url = "https://min-api.cryptocompare.com/data/all/coinlist"
r = requests.get(url)
crypto_data = r.json()['Data']
```

```python
# Create a DataFrame 
# HINT: You will need to use the 'Data' key from the json response, then transpose the DataFrame.
crypto_df = pd.DataFrame(crypto_data).T
crypto_df.head()
```

### Data Preprocessing

```python
# Keep only necessary columns:
# 'CoinName','Algorithm','IsTrading','ProofType','TotalCoinsMined','TotalCoinSupply'
crypto_df = crypto_df[['CoinName','Algorithm','IsTrading','ProofType','TotalCoinsMined','MaxSupply']]
```

```python
# Keep only cryptocurrencies that are trading
crypto_df = crypto_df[crypto_df.IsTrading.eq(True)]
crypto_df
```

```python
# Keep only cryptocurrencies with a working algorithm
crypto_df = crypto_df[crypto_df.Algorithm != "N/A"]
crypto_df
```

```python
# Remove the "IsTrading" column
crypto_df.drop(["IsTrading"], axis = 1, inplace = True)
crypto_df
```

```python
# Remove rows with at least 1 null value
crypto_df.dropna(inplace=True)
```

```python
# Remove rows with cryptocurrencies having no coins mined
crypto_df = crypto_df[crypto_df.TotalCoinsMined != 0]
crypto_df
```

```python
# Drop rows where there are 'N/A' text values
crypto_df = crypto_df[crypto_df.ProofType != "N/A"]
crypto_df
```

```python
# Store the 'CoinName'column in its own DataFrame prior to dropping it from crypto_df
coinname_df = crypto_df.iloc[:,0]
coinname_df
```

```python
# Drop the 'CoinName' column since it's not going to be used on the clustering algorithm
clean_crypto_df = crypto_df.drop(columns="CoinName")
clean_crypto_df
```

```python
# Create dummy variables for text features
X = pd.get_dummies(clean_crypto_df, columns = ["Algorithm", "ProofType"])
X.head()
```

```python
# Standardize data
s = StandardScaler()
s.fit(X)
X_scaled = s.transform(X)
X_scaled[:1]
```

### Reducing Dimensions Using PCA

```python
# Use PCA to reduce dimensions to 3 principal components
pca = PCA(n_components=3)
crypto_pca = pca.fit_transform(X_scaled)
```

```python
# Create a DataFrame with the principal components data
df_crypto_pca = pd.DataFrame(data = crypto_pca, columns = ["PC1", "PC2", "PC3"], index= clean_crypto_df.index)
df_crypto_pca
```

### Clustering Crytocurrencies Using K-Means

#### Find the Best Value for `k` Using the Elbow Curve

```python
inertia = []
k = list(range(1, 11))

# Calculate the inertia for the range of k values
# Finding the best k
for i in k:
    km = KMeans(n_clusters=i, random_state=0)
    km.fit(df_crypto_pca)
    inertia.append(km.inertia_)

# Creating dataframe to plot Elbow Curve with hvPlot

elbow_data = {"k": k, "inertia": inertia}
elbow_df = pd.DataFrame(elbow_data)

# Create the Elbow Curve using hvPlot
elbow_df.hvplot.line(x = "k", y = "inertia", title = "Elbow Curve", xticks = k)
```

Running K-Means with `k=<your best value for k here>`

```python
# Initialize the K-Means model
crypto_model = KMeans(n_clusters = 5, random_state = 5)
# Fit the model
crypto_model.fit(df_crypto_pca)
# Predict clusters
predictions = crypto_model.predict(df_crypto_pca)
# Create a new DataFrame including predicted clusters and cryptocurrencies features
crypto_clusters_df = pd.concat([crypto_df, df_crypto_pca], axis = 1)
crypto_clusters_df["Class"] = crypto_model.labels_
crypto_clusters_df
```

### Visualizing Results

#### Scatter Plot with Tradable Cryptocurrencies

```python
# Scale data to create the scatter plot
scaled_data = MinMaxScaler().fit_transform(crypto_clusters_df[['TotalCoinsMined', 'MaxSupply']])
scaled_df = pd.DataFrame(scaled_data, columns=['TotalCoinsMined', 'MaxSupply'], index=crypto_clusters_df.index)
scaled_df['Class'] = crypto_clusters_df['Class']
scaled_df
```

```python
# Plot the scatter with x="TotalCoinsMined" and y="TotalCoinSupply"
scatter_plot = scaled_df.hvplot.scatter(x='TotalCoinsMined', y='MaxSupply',
                                        hover_cols=['Class'], by=['Class'])
scatter_plot
```

#### Table of Tradable Cryptocurrencies

```python
# Table with tradable cryptos
tradable_crypto = crypto_clusters_df[['CoinName', 'Algorithm', 'ProofType', 'TotalCoinsMined', 'MaxSupply', 'Class']]
tradable_crypto.hvplot.table()
```
